# index yerindeki 0 1 2 i yerine yazı yazma

```python
student = pd.Series(data = ["elina",20,"CE"], index = ["name","age","department"])
print(student)
print("-------")
print(student.values) #değişkeni yani ['elina' 20 'CE'] i yazar
print("-------")
print(student.index)# indexi yaniIndex(['name', 'age', 'department'] i yazar
print("-------")
print(student.shape)# 3,1 lik matrisi (3,)tuple olarak yazar
print("-------")
print(student.ndim)# boyutunu yazar yani 1 yazar
print("-------")
print(student.dtype)#object olarak yazar
print("-------")
print(student.size)#eleman sayısını yazar yani 3 elemanlıdır
print("-------")
print(student.age)# age i yani 20 yazar
```

```python
numbers = pd.Series([4, 2.5, 2])
print(numbers)
print("------")
print(numbers.sum())# toplamını yazar
print("------")
print(numbers.mean())#ortalamasını yazar
print("------")
print(numbers.product())#eleamnalrın çarpımını yazar
```

# loc --> location, iloc --> index location

* loc kullandığımızda parametre olarak index deki değeri yazarız.Mesela:

```python
student = pd.Series(["elina","elinson",20,"CE"],["name","surname","age","department"])
print(student)
print("-------")
print(student.loc["name"])
print(student.loc["age"])
```

* iloc kullandığımızda parametre olarak index sayısı yazılır.Mesela:

```python
student = pd.Series(["elina","elinson",20,"CE"],["name","surname","age","department"])
print(student)
print("-------")
print(student.iloc[1])
print(student.iloc[3])
```

# Eleman Silme

```python
student = pd.Series(["elina","elinson",20,"CE"],["name","surname","age","department"])
print(student)
print("-------")
student.drop("name")
print(student)
```

* Böyle yaparsak orijinal veride değişiklik yapmayız.Böyle olsaydı bir veri kolay kolay silinebilirdi.Bu yüzden aşağıdaki gibi yapmak lazım.

```python
student = pd.Series(["elina","elinson",20,"CE"],["name","surname","age","department"])
print(student)
print("-------")
student2 = student.drop("name")
print(student2)
```

```python
student = pd.Series(["elina","elinson",20,"CE"],["name","surname","age","department"])
print(student)
print("-------")
student.drop("name",inplace = True)
print(student)
```

* Ama illaki o veriyi silmek istiyorsak, böyle yaparız. Yani inplace metodunu kullanırız. 

# len() kullanmak ve seriesi liste ve dictinary yapmak.

```python
student = pd.Series(["elina","elinson",20,"CE"],["name","surname","age","department"])
print(student)
print("-------")
print(len(student))
print("-------")
print(list(student))
print("-------")
print(dict(student))
```

# max(), min()

```python
student = pd.Series(["elina","elinson",20,"CE"],["name","surname","age","department"])
print(student)
print("-------")
print(max(student))
```

* Burada yazmadı ama derleyicide yazarsak,en sondakini yazar yani "CE" yazar

```python
student = pd.Series(["elina","elinson",20,"CE"],["name","surname","age","department"])
print(student)
print("-------")
print(min(student))
```
