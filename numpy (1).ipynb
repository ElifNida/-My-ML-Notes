# np.stack()

> **1 --> np.vstack() --> vertical stack yani dikey iliştrme**

```python
x = np.arange(3)
print(x)
print("--------")
y = np.arange(9).reshape(3,3)
print(y)
print("---------")
z = np.vstack((x,y))
print(z)
```

* x i y nin üzerine yazar

> **2 --> np.hstack() --> hortizontal stack yani yatay iliştrme**

```python
x = np.arange(3)
print(x)
print("--------")
y = np.arange(9).reshape(9,)
print(y)
print("---------")
z = np.hstack((x,y))
print(z)
```

* Boyutlarının aynı olamsı lazım o yüzden reshape(9,) yaptık.

# slicing yani ndArray in alt kümesine ulaşma Tek Boyutluda

```python
x = np.arange(10)
print(x)
print("------")
print(x[2:6]) # 2den 6. elemana kadar yazar. 6 dahil değil
print("------")
print(x[4:])#4 ten en sona kadar yazar.
print("------")
print(x[:7])#en baştan 7.elemana kadar yazar. 7 dahil değil
```

```python
x = np.arange(10)
print(x)
print("------")
print(x[:-1]) # -1 demek son eleman demek, son elemanı yazmaz.
print("------")
print(x[-4:-2])
```

# slicing yani ndArray in alt kümesine ulaşma 2 Boyutluda

```python
x = np.arange(25).reshape(5,5)
print(x)
print("--------")
y = x[0:3, 1:4] #0 1 2. satırın , 1 2 3. sütunu yazar
print(y)
print("--------")
z = x[2:]
print(z)# 2. satırdan başlayarak bütün satırları yazar
print("--------")
t = x[::]
print(t)# x matrisinin hepsini yazar
print("--------")
v = x[:, 3:]
print(v)#bütün satırların 3. sütunundan itibaren yazar.
print(np.may_share_memory(t,v))# t ve v nin sadece sayıları değişiyor,kapladığı yerler aynı
```

```python
x = np.arange(9).reshape(3,3)
print(x)
x[0,0] = 100
print(x)
```

# copy()

```python
x = np.arange(16).reshape(4,4)
print(x)
print("---------")
y = np.copy(x[0:3, 1:3])
print(y)
print(np.shares_memory(x,y))#slicing gibi kapladığı yer aynı değil. np.copy() nin tuttuğu yerde farklıdır.
```
