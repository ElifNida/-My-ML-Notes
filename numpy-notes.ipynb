# Grup İşlemleri ve Sıralama

# np.intersect1d() , np.setdiff1d(), np.union1d(), np.in1d()

* Grup işlemleri, genellikle 2 tane ndAarray ı karşilaştırırken kullanılır.
* Aşağıdaki işlemler sıralı bir şekilde gerçekleştirilir.

```python
x = np.array([2,5,3,6,9])
y = np.array([6,1,8,7,3])

z =np.intersect1d(x,y)# x ve ye de ortak olan elemanları yazar, yani x ve y
print(z)
print(type(z))
print("--------")

t = np.setdiff1d(x,y)# x te olan ama y de olmayan sayıları yazar.
print(t)
print("--------")
g = np.union1d(x,y)# x ve ye olan bütün elemanalrı yazar. ikisindede olan elemanları 2 kere yazmaz sadece bir kere yazar
print(g)
print("--------")
h = np.in1d(x,y)# x te olan elemanalr y de de var mı diye bakar ve ona göre true ya da false yazar.
print(h)
```

* np.intersect1d(x,y) --> x ve y de ortak elemanlar yazar
* np.setdiff1d(x,y) --> x te olan ama y de olmayan sayıları yazar
* np.union1d(x,y) --> x ve y deki elemanalrı yazar.ortak olan elemanları sadece 1 kere yazar
* np.in1d(x,y) -->  x te olan elemanalar y de de var mı diye bakar ve ona göre true ya da false yazar.
